## **5. Descriptive Statistics**

Calculate central tendency (Mean, Median, Mode) and dispersion (Min, Max, Standard Deviation) measures.

```python
# Complete Descriptive Statistics using describe()
print("📈 Complete Descriptive Statistics:")
print("=" * 80)
df_cleaned.describe()
```

## 2️⃣ Central Tendency Analysis
> 📌 *Calculate Mean, Median and Mode for Age, Salary and Health Scores*

---

```python
# Calculate Mean, Median, and Mode for numerical columns
numerical_columns = ['Age', 'Healthy_Eating_Score', 'Active_Lifestyle_Score', 'Salary']

print("📊 CENTRAL TENDENCY MEASURES")
print("=" * 70)

for col in numerical_columns:
    mean_val = df_cleaned[col].mean()
    median_val = df_cleaned[col].median()
    mode_val = df_cleaned[col].mode()[0]
    
    print(f"\n🔹 {col}:")
    print(f"   Mean   : {mean_val:.2f}")
    print(f"   Median : {median_val:.2f}")
    print(f"   Mode   : {mode_val:.2f}")
```

---

## 3️⃣ Dispersion Analysis
> 📌 *Measure spread using Standard Deviation, Variance, Range and IQR*

---

```python
# Calculate Dispersion Measures
print("📊 DISPERSION MEASURES")
print("=" * 70)

for col in numerical_columns:
    min_val = df_cleaned[col].min()
    max_val = df_cleaned[col].max()
    std_val = df_cleaned[col].std()
    var_val = df_cleaned[col].var()
    range_val = max_val - min_val
    
    print(f"\n🔹 {col}:")
    print(f"   Minimum          : {min_val:.2f}")
    print(f"   Maximum          : {max_val:.2f}")
    print(f"   Range            : {range_val:.2f}")
    print(f"   Standard Deviation: {std_val:.2f}")
    print(f"   Variance         : {var_val:.2f}")
```

```python
# Create a Summary Table for All Statistics
stats_summary = pd.DataFrame({
    'Mean': df_cleaned[numerical_columns].mean(),
    'Median': df_cleaned[numerical_columns].median(),
    'Mode': df_cleaned[numerical_columns].mode().iloc[0],
    'Min': df_cleaned[numerical_columns].min(),
    'Max': df_cleaned[numerical_columns].max(),
    'Std Dev': df_cleaned[numerical_columns].std(),
    'Variance': df_cleaned[numerical_columns].var()
}).round(2)

print("📋 COMPLETE STATISTICS SUMMARY TABLE")
print("=" * 80)
stats_summary
```

```python
# Group-wise Mean Statistics
print("📊 GROUP-WISE MEAN STATISTICS")
print("=" * 70)
group_mean = df_cleaned.groupby('Blood_Group')[numerical_columns].mean().round(2)
group_mean
```

## **7. Interpretation**

Explain the results in simple words and determine whether mean or median is more representative.

```python
# Check Skewness to Determine Best Central Tendency Measure
print("📊 SKEWNESS ANALYSIS")
print("=" * 70)
print("\n🔹 Skewness helps determine if Mean or Median is more representative:")
print("   • Skewness ≈ 0: Symmetric → Mean is good")
print("   • Skewness > 0: Right-skewed → Median is better")
print("   • Skewness < 0: Left-skewed → Median is better")
print("\n" + "=" * 70)

for col in numerical_columns:
    skew_val = df_cleaned[col].skew()
    mean_val = df_cleaned[col].mean()
    median_val = df_cleaned[col].median()
    
    if abs(skew_val) < 0.5:
        recommendation = "Mean (data is approximately symmetric)"
    else:
        recommendation = "Median (data is skewed)"
    
    print(f"\n🔹 {col}:")
    print(f"   Skewness: {skew_val:.3f}")
    print(f"   Mean: {mean_val:.2f} | Median: {median_val:.2f}")
    print(f"   ✅ Recommended: {recommendation}")
```

### **Interpretation in Simple Words:**

**Age:**
- The average employee age gives us an idea of the workforce maturity
- If mean ≈ median, the age distribution is balanced

**Salary:**
- Salary often shows right-skewness (few high earners pull the mean up)
- In such cases, **median salary** better represents typical employee earnings

**Healthy Eating & Active Lifestyle Scores:**
- These scores indicate employee wellness levels
- Higher scores suggest healthier lifestyle choices

## 6️⃣ Outlier Detection
> 📌 *Identify anomalies using Z-Score method and IQR fencing*

---

```python
# Skewness & Kurtosis
from scipy import stats
print("Age Skewness:", stats.skew(df_cleaned['Age']))
print("Age Kurtosis:", stats.kurtosis(df_cleaned['Age']))
print("Salary Skewness:", stats.skew(df_cleaned['Salary']))

# Quartiles & IQR
Q1 = df_cleaned['Salary'].quantile(0.25)
Q3 = df_cleaned['Salary'].quantile(0.75)
IQR = Q3 - Q1
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")

# Coefficient of Variation (CV)
cv_salary = (df_cleaned['Salary'].std() / df_cleaned['Salary'].mean()) * 100
print(f"Salary CV: {cv_salary:.2f}%")

# Percentiles
print(df_cleaned['Age'].describe(percentiles=[.10,.25,.50,.75,.90]))

# Z-Score (outlier detection)
df_cleaned['Age_Zscore'] = stats.zscore(df_cleaned['Age'])
df_cleaned['Salary_Zscore'] = stats.zscore(df_cleaned['Salary'])
outliers = df_cleaned[(df_cleaned['Age_Zscore'].abs() > 2) | 
              (df_cleaned['Salary_Zscore'].abs() > 2)]
print(f"Outliers found: {len(outliers)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histogram with KDE (better than plain bar)
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, col in enumerate(['Age', 'Salary', 
                          'Healthy_Eating_Score']):
    axes[i].hist(df_cleaned[col], bins=10, 
                 edgecolor='black', alpha=0.7)
    axes[i].set_title(f'{col} Distribution')
plt.tight_layout()
plt.show()

# 2. Correlation Heatmap
numeric_cols = df_cleaned.select_dtypes(include='number')
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_cols.corr(), 
            annot=True, fmt='.2f',
            cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()

# 3. Salary by Blood Group (grouped analysis)
plt.figure(figsize=(8, 5))
sns.boxplot(x='Blood_Group', y='Salary', data=df_cleaned)
plt.title('Salary Distribution by Blood Group')
plt.show()

# 4. Scatter: Age vs Salary
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Salary', 
                hue='Blood_Group', data=df_cleaned, s=100)
plt.title('Age vs Salary by Blood Group')
plt.show()

# 5. Health Score vs Active Lifestyle
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Healthy_Eating_Score', 
                y='Active_Lifestyle_Score',
                hue='Blood_Group', data=df_cleaned, s=100)
plt.title('Health Eating vs Active Lifestyle')
plt.show()
```

```python
# Group stats by Blood Group:
group_stats = df_cleaned.groupby('Blood_Group').agg({
    'Age':    ['mean', 'median', 'std'],
    'Salary': ['mean', 'median', 'std', 'min', 'max'],
    'Healthy_Eating_Score':  'mean',
    'Active_Lifestyle_Score': 'mean'
}).round(2)

print("=== Stats by Blood Group ===")
print(group_stats)

# High vs Low salary earners
median_salary = df_cleaned['Salary'].median()
df_cleaned['Salary_Category'] = df_cleaned['Salary'].apply(
    lambda x: 'High' if x > median_salary else 'Low'
)
print(df_cleaned.groupby('Salary_Category')[
    ['Age', 'Healthy_Eating_Score', 
     'Active_Lifestyle_Score']
].mean())
```
